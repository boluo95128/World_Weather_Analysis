```python
trip_start=preferred_city_data_df.loc[preferred_city_data_df["City"]=="Tecoanapa"]
trip_stop_1=preferred_city_data_df.loc[preferred_city_data_df["City"]=="Coahuayana"]
trip_stop_2=preferred_city_data_df.loc[preferred_city_data_df["City"]=="Ixtapa"]
trip_end=preferred_city_data_df.loc[preferred_city_data_df["City"]=="Valle De Banderas"]
```

```python
start=trip_start["Lat"].values[0], trip_start["Lng"].values[0]
stop1=trip_stop_1["Lat"].values[0], trip_stop_1["Lng"].values[0]
stop2=trip_stop_2["Lat"].values[0], trip_stop_2["Lng"].values[0]
end=trip_end["Lat"].values[0], trip_end["Lng"].values[0]
print(start)
print(stop1)
print(stop2)
print(end)
```

```python
# To create a route map for the trip

fig=gmaps.figure()
suggested_route=gmaps.directions_layer(start, end, waypoints=[stop1, stop2], travel_mode="DRIVING")
fig.add_layer(suggested_route)
fig
```
